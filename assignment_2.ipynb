```python
#Question 1:
import numpy as np
n1=np.random.rand(10,1)
print(n1)
mean=np.mean(n1)
print("Mean =" ,mean)
```

```python
#Question 2:
n2=np.random.rand(20,1)
print(n2)
standev=np.std(n2)
print("Standard Deviation =" ,standev)
variance=np.var(n2)
print("Variance =" ,variance)
```

```python
#Question 3:
A=np.random.rand(10,20)
B=np.random.rand(20,25)
M=np.matmul(A,B)
print("After Multiplication =\n",M)
print(M.shape)
sum=np.sum(M)
print("Sum is =", sum)
```

```python
#Question 4:
import math
A=np.random.rand(10,1)
for x in np.nditer(A,op_flags=['readwrite']):
    x[...]=1/(1+math.exp(-x))
print("NEW ARRAY",A)
print(A.shape)
    
```
