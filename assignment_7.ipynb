## UNIVARIATE ANALYSIS
uni=one/single variate=variable

It will help us in finding outliers

Various plots in this are scatter,histogram,kde,pdf,etc

# 1. Fixed Acidity plot

```python
#1.
sns.distplot(data['fixed acidity'],bins=50)
plt.show()
```

### Analysis:
1)it is a normal distribution.

2)maximum sample is in between 6-8.

3)outliers are present at the last between 10-14.

## 2. Volatile Acidity plot

```python
plt.hist(x=data['volatile acidity'],bins=50)
plt.grid()
plt.title('Histogram of Volatile acidity in white wine')
```

### Analysis:
1)it is both normal and slightly skewed distribution.

2)most sample has volatile acidity in between 0.2-0.4.

3)outliers are present at the last between 0.6-0.8.

## 3. Citric Acid Plot

```python
sns.kdeplot(data['citric acid'], shade=True)
plt.show()
```

### Analysis:
1)it is both normal distribution and slightly skewed at the last.

2)most sample have citric acid content in between 0.0-0.5.

3)outliers are present in less number.

## 4. Alcohol plot

```python
plt.hist(x=data['alcohol'])
plt.grid()
```

### Analysis:
1)it is a right skewed distribution.

2)most sample have alcohol content in them and it ranges in between 9-12.

3)less outliers.

## 5. Density plot

```python
sns.distplot(data['density'])
plt.title("Density distplot")
plt.show()
```

### Analysis :

1.)graph is skewed.

3.)Many outliers are present.


## 6. Sulphates plot

## 7. pH plot

### Analysis :

3.)Many outliers are present.

### Analysis :

## BIVARIATE ANALYSIS
bi=two variate=variable

it gives more detailed analysis

plots include jointplot,regplot,boxplot,etc

## 9. quality vs alcohol

```python
sns.jointplot(x='quality', y='alcohol', data=data)
plt.show()
```

### Analysis 
1.) Good quality wines sample are less and have high concentrationof alcohol in them.

2.) Most of the samples have quality grade of 5,6,7.

3.) Major sample have concentration in range of 9-13.

## 10. alcohol vs density

```python
sns.regplot(x='alcohol', y='density', data=data)
plt.show()
```

### Analysis:
Density and alcohol are correlated.

They are inversely proportional.

Increase in concentration leads to decrease in alcohol concentration.

There are some outlier.

## 11. quality vs density

```python
sns.regplot(x='quality' , y='density', data=data)
plt.show()
```

### Analysis: 
1.)As density increases, quality decreases.

2.)an important point for analysis.

## 12. Quality vs residual sugar

```python
sns.boxplot(x='quality',y='residual sugar',data=data)
plt.title("Quality and pH residual sugar")
plt.show()
```

### Analysis: 
Residual sugar content of good quality wine is very less.

the average quality wines has residual sugar contents between 0-20.

## 13. quality vs pH

```python
sns.regplot(x='quality',y='pH',data=data)
plt.title("Quality  and pH regplot")
plt.show()
```

### Analysis:
There is correlation between ph and quality.

pH's for most wines lies between 2.8-3.8.

## 14. free vs total sulfur dioxide

```python
sns.regplot(x='free sulfur dioxide',y='total sulfur dioxide',data=data,scatter=True,color='orange')
plt.show()
```

## 15. quality vs density

```python
sns.pointplot(x='quality',y='density',data=data)
plt.title("Quality and Density pointplot")
plt.show()
```

### Analysis:
as density increases,quality decreases.

## MULTIVARIATE ANALYSIS
analysis on more than two variables.

## 16.

```python
grid = sns.FacetGrid(data, col='quality',col_wrap=3)
grid.map(plt.scatter,'pH','residual sugar')
plt.show()
```

## 17.

```python
grid=  sns.FacetGrid(data, col='quality',col_wrap = 4)
grid.map(plt.scatter,'alcohol','density',alpha = 0.4)
```

## 18.

```python
sns.lmplot("fixed acidity", "volatile acidity", data,hue="quality", fit_reg=False)
plt.show()
```

## 19.

```python
sns.lmplot("density", "pH", data,hue="quality", fit_reg=False)
plt.show()
```
